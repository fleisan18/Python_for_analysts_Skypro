### Индекс

Индекс — это названия строк в Pandas, они нужны, чтобы идентифицировать каждую строчку.

В нашем датафрейме индекс — это автоматически заданная последовательность чисел от 0 до 104693. 

У нас есть уникальные идентификаторы пользователей, удобно будет использовать именно их в качестве индекса. 

```python
workouts.index
```

### `.set_index()`

Есть несколько способов переопределить индекс, например, метод `.set_index()`

Этот метод принимает два аргумента:

1. Имя столбца
2. Параметр `inplace`, который влияет на то, изменяем ли мы изначальный датафрейм или нет. 

Если `inplace = True`, наш датасет изменяется, если мы проигнорируем этот аргумент, по умолчанию он равен `False` и датафрейм не перезаписывается. 

```python
workouts.set_index('workout_id')

# или тот же результат даст код
workouts.set_index('workout_id', inplace=False)
```

```python
workouts.head(2) # датасет не изменился
```

```python
workouts.set_index('workout_id',  inplace=True) # перезаписали датасет
```

```python
workouts.head() # проверим, что теперь в индексах
```

### `.reset_index()`

Если мы хотим вернуть колонку, которую отправили в индекс, мы можем сделать это с помощью `.reset_index()`. Значения индекса станут колонкой, а в индексах строк появится последовательность от 0 с шагом в единицу.

```python
workouts.reset_index() # это действие существует только в рамках данной ячейки
```

В работе с pandas сбрасывать индекс часто приходится, когда в процессе обработки данных мы получаем датафрейм с одним столбцом. Такой формат называется серии или `Series`. Каждый датафрейм — это набор столбцов, набор серий.

Часть методов пандаса не работают с сериями, поэтому мы, приходя к ним в работе, сбрасываем индекс с помощью `.reset_index()` или каким-то другим способом.

```python
users['workouts_total'] # мы обратились к одной колонке и получили объект Series
```

```python
type(users['workouts_total'])
```

```python
users['workouts_total'].reset_index()
# вызвав метод reset_index() мы отправляем наш текущий индекс в еще одну колонку и получаем DataFrame
```

```python
# Еще один способ получить DataFrame из Series -—явно передать Series в конструктор объекта DataFrame
pd.DataFrame(users['workouts_total'])
```

### `.groupby( ) `


Позволяет для любой колонки агрегировать данные разными методами группировки. То есть мы берем колонку и для каждого уникального значения из нее выводим агрегат по какой-то другой колонке, используя методы группировки. 

   - `.count()` — подсчет количества вхождений строк

### Группировка по месяцу `.dt.month`

Можно группировать по датам данные, для этого просто нужно добавить в аргумент `groupby` столбец датафрейма и метод `.dt.month`

Так мы можем группировать данные по:/
- месяцам — `month`

- годам — `year`

- дням — `day`

Cгруппируем по месяцу и определим, сколько клиентов пришли на тренировки помесячно.

## Соединение нескольких таблиц в одну

##  `.merge()`


Аналогично JOIN в SQL

Часто мы получаем данные из разных источников, с разными комплектами данных, которые имеют пересечения, например, данные по оплатам пользователей и данные по урокам пользователей

Чтобы информацию из одного источника сопоставлять с информацией из другого, нам нужно их объединять.

**Для объединения данных в pandas существует функция `.merge()`.**

Он позволяет объединить данные по общему ключу. Например, во всех наших датасетах есть `id` клиента, он может быть ключом.

Применим функцию `.merge()`, передав в него два датафрейма
На выходе получим новый датафрейм, который содержит в себе данные из первого и второго датафрейма

```python
pd.merge(payments, users)
```


**В Pandas cуществуют разные типы соединений:**

1. **inner** — работает по умолчанию и выводит значения, ключ по которым встречается и в первой, и во второй таблице 

2. **left** — выводит все значения из первой таблицы и значения из второй таблицы, к которым нашлись ключи из первой

3. **right** — НИКОГДА НЕ ПОЛЬЗУЕМСЯ RIGHT JOIN

4. **outer** — выводит все значения из обеих таблиц и совпадающие ключи и уникальные для первой или второй таблицы.

Для всех отсутствующих значений присваивается `NaN`.

Передать тип соединения нужно через аргумент `how='тип соединения'`

### Посмотрим, как объединятся другие таблицы

```python
pd.merge(workouts, users, how='left') # попробуем смержить пользователей и уроки с помощью left join
```

Код выдает ошибку, потому что в наших данных pandas не нашел колонок, по которым мы хотели соединить данные. 

**Давайте напрямую передадим колонки, по которым мы хотим объединить таблицы.**

Это делается с помощью еще двух аргументов:

1. `left_on='имя колонки первой таблицы'`
2. `right_on='имя колонки второй таблицы'`



Добавим в аргументы left_on и right_on названия столбцов для объединения



```python
workouts.info() # вспомним, как называется колонка с айдишниками клиентов в первой таблице
```

```python
users.info() # вспомним, как называется колонка с айдишниками клиентов во второй таблице
```

```python
workouts_users =  pd.merge(workouts, users, # создадим новый датафрейм
         how='left',                      
         left_on='client_id', right_on='user_id') # здесь дополним аргументы
```

```python
workouts_users.head()
```

```python
workouts_users.info() # посмотрим сколько у нас пропусков в данных
```

**outer join — опасная странная штука, потому что у вас нарушается уровень детализации таблицы.**

## Сортировка

### `sort_values()`

- При сортировке датафрейма обязательный аргумент — это `by=`: по какой колонке сортировать. 
- Сортировка понимает все типы данных
- Для сортировки по индексам есть `.sort_index()`

```python
user_info.sort_values(by='date_diff') # отсортируем пользователей по сроку между первым уроком и первой оплатой 
```

Можно сортировать по нескольким колонкам сразу, при этом порядок колонок в списке определяет порядок сортировки

```python
user_info.sort_values(by=['date_diff','start_at']) 
```

По умолчанию сортировка выполняется по возрастанию, чтобы сделать по убыванию, нужно добавить условие в аргументе `ascending=False`

```python
user_info.sort_values(by='date_diff',
                ascending=False)
```

### `.apply()`

Позволяет нам применять функции ко всем значениям колонки, которую мы укажем.

### Задача:

Мы хотим создать новый столбец, который промаркерует всех пользователей как `Быстрый`, если с даты вводной тренировки до даты оплаты прошло меньше времени, чем в среднем, и `Медленный`в противном случае.


```python
# Среднее количество дней от вводного занятия до платежа
mean_value = user_info['date_diff'].mean()
mean_value
```

```python
def checher(datediff_in_days):
    if datediff_in_days <= mean_value:
        return 'Быстрый'
    else:
        return 'Медленный'
```

```python
#используем нашу функцию
user_info['client_type'] = user_info['date_diff'].apply(checher)
```
